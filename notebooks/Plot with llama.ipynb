```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding

embed_model = HuggingFaceEmbedding(model_name="BAAI/bge-small-en-v1.5")
```

```python
sentence = "This is a test sentence."
sentence_embedding = embed_model.get_text_embedding(sentence)

print(f"Embedding for '{sentence}': {sentence_embedding}")
```

```python
import re
import pandas as pd
```

```python
df = pd.read_csv('/content/drive/My Drive/data/cosmetic.csv')
df
```

```python
unique_ingredients = set()

for ingredient_list in df['ingredients']:
  ingredients = ingredient_list.split(",")
  unique_ingredients.update(ingredient.strip() for ingredient in ingredients)
```

```python
import matplotlib.pyplot as plt

embeddings = []
ingredients_list = list(unique_ingredients)[:100]
for a in ingredients_list:
  embeddings.append(embed_model.get_text_embedding(a))
```

```python
from sklearn.decomposition import PCA
from mpl_toolkits.mplot3d import Axes3D
import plotly.express as px

additional_words = ['water', 'oil', 'dog', 'pharmaceutical']

embeddings_to_plot = embeddings[:20]
embedding_labels = ingredients_list[:20]

for word in additional_words:
  embeddings_to_plot.append(embed_model.get_text_embedding(word))
  embedding_labels.append(word)


pca = PCA(n_components=3)
embeddings_3d = pca.fit_transform(embeddings_to_plot)

fig = px.scatter_3d(x=embeddings_3d[:, 0], y=embeddings_3d[:, 1], z=embeddings_3d[:, 2],
                    text=embedding_labels, size= [10] * len(embedding_labels))

fig.update_traces(marker=dict(size=5))  # Reduce marker size slightly
fig.show()
```
